### 처리 결과 확인

```python
datas = [seoul_show, daegu_show, seoul_yanolza, daegu_yanolza]
for data in datas :
    print(data.info(), '\n')
```

```python
def detail_df(datas):
    df = pd.DataFrame({
                        'shape':[data.shape for data in datas],
                        'columns':[data.columns for data in datas],
                        "na sum":[data.isna().sum().sum() for data in datas]},
                        index = ['서울 공공데이터', '대구 공공데이터', '서울 야놀자', '대구 야놀자']
                        )
    return df
    
detail_df(datas)
```

### 결론
- 공공 데이터 내에서 
    - 문화/ 공연이 시행 횟수 자체는 서울이 대구보다 많지만,
    - 인구수 비례 시행 횟수는 유사하거나 대구가 더 많다.


- 공공 데이터와 야놀자 상품 목록 통합 데이터 에서
    - 전체적인 공연 시행 수가 서울이 압도적으로 많았고
    - 인구 수 비례 시행 정도도 서울이 높았다.
    - 공공데이터 내의 자료는 대구가 더욱 많지만, 야놀자 어플 상품의 수가 서울 내에 매우 많아 이런 결과를 보였다.


- 구별 분석에서
    - 대구 내에서는 공연이 중구와 달서구에 집중되어 있고
    - 20년 12월 이전에는 중구, 이후는 달서구에 많은 공연이 있었다.
    - 서울 내에서는 마포구에 집중되어 있다.


#### 부족한 점
- 공공 데이터 자료에서는 지역의 구별로 시행내역이 저장 되어 있어 각 데이터의 기간이 일정하지 않았다.
- 어떤 자료는 2022년의 자료만 있는것도 존재 하였다.
