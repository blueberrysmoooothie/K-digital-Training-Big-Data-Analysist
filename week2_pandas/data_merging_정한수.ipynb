```python
import pandas as pd
import os
DEAGU_DIR_PATH = './대구_공연행사/'
SEOUL_DIR_PATH = './서울_공연행사/'
DIR_PATH = './공연/'
```

#### 공공 데이터 통합 문서로 변환하기

```python
deagu_show = pd.read_csv(DEAGU_DIR_PATH + os.listdir(DEAGU_DIR_PATH)[0] )
seoul_show = pd.read_csv(SEOUL_DIR_PATH + os.listdir(SEOUL_DIR_PATH)[0] )
```

```python

deagu_show_total = pd.DataFrame(columns = deagu_show.columns)
for file in os.listdir(DEAGU_DIR_PATH):
    deagu_show = pd.read_csv(DEAGU_DIR_PATH + file )
    deagu_show_total = pd.concat([deagu_show_total, deagu_show], ignore_index= True)


seoul_show_total = pd.DataFrame(columns = seoul_show.columns)   
for file in os.listdir(SEOUL_DIR_PATH):
    seoul_show = pd.read_csv(SEOUL_DIR_PATH + file )
    seoul_show_total = pd.concat([seoul_show_total, seoul_show], ignore_index= True)
```

```python
deagu_show_total.info()
```

```python
seoul_show_total.info()
```

```python
def df_detail(*datas):
    df = pd.DataFrame({
        'shape':[data.shape for data in  datas],
        'columns':[data.columns for data in  datas],
        "na sum":[data.isna().sum().sum() for data in  datas]},
        index = ['서울', '대구']
        )
    return df
df_detail(seoul_show_total, deagu_show_total)
```

```python
deagu_show_total.to_csv('deagu_show.csv',index= False)
seoul_show_total.to_csv('seoul_show.csv',index= False)
```

#### 야놀자 데이터 통합 문서로 변환하기

```python
os.listdir(DIR_PATH)
```

```python
show2008 = pd.read_csv(DIR_PATH + os.listdir(DIR_PATH)[0])
show2008.info()
```

```python
deagu = pd.DataFrame(columns = show2008.columns)
seoul = pd.DataFrame(columns = show2008.columns)

for file in os.listdir(DIR_PATH):
    month_data = pd.read_csv(DIR_PATH + file)
    month_deagu = month_data[[('대구' in x) and (not '부산' in x) for x in month_data['AREA_NM']]]
    month_seoul = month_data[['서울' in x for x in month_data['AREA_NM']]]
    deagu = pd.concat([deagu, month_deagu])
    seoul = pd.concat([seoul, month_seoul])
```

```python
seoul.info()
```

```python
df_detail(seoul, deagu)
```

```python
deagu.to_csv('deagu_yanolza.csv',index= False)
seoul.to_csv('seoul_yanolza.csv',index= False)
```
